```python
msk = np.random.rand(len(dataset)) < 0.8

train = dataset[msk]
test = dataset[~msk]
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn import linear_model
model= linear_model.LinearRegression()
train_x = np.asanyarray(train[['Confirmed']])  #numpy.asanyarray()function is used when we want to convert input to an array
train_y = np.asanyarray(train[['Cured']])
model.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', model.coef_)
print ('Intercept: ',model.intercept_)
```

```python
plt.scatter(train.Confirmed, train.Cured, color='royalblue')
plt.plot(train_x,model.coef_[0]*train_x + model.intercept_[0],color='RED')
plt.ylabel("CURED")
plt.xlabel("CONFIRMED CASES")
```
