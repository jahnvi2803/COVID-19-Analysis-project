```python
#IMPORTING REQUIRED LIBRARIES
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
from matplotlib.colors import Normalize
from numpy.random import rand
```

```python
#for datewise data of no. cases of the state
df1=pd.read_csv('UP DAILY COVID DATA.csv')
```

```python
df1.head()
```

```python
df1.info()
```

```python
#plot for analysing day by day increase or decrease in no. of cases

plt.xlabel('Date')
plt.ylabel('cases')
ttl1=plt.title('DATEWISE ANALYSIS : ')
ttl1.set_position([.5, 1.05])

plt.plot('DATE','CONFIRMED',data=df1,color='orange',label='confirmed',linestyle='-', marker='o')
plt.plot('DATE','RECOVERED',data=df1,color='green',label='recovered',linestyle='-', marker='o')
plt.plot('DATE','DECEASED',data=df1,color='red',label='deceased',linestyle='-', marker='o')

plt.xticks(rotation='vertical')
plt.subplots_adjust(-1)
plt.legend()
plt.show()
```

```python
#pie chart giving full insights of cases in UP till now

labels = 'Confirmed', 'Active','Recovered', 'Deceased'
sizes = [45163,16445,27634,1084]
colors = ['gold', 'deepskyblue', 'limegreen', 'crimson']
explode = (0,0.2,0,0)  

ttl=plt.title('OVERALL INSIGHTS OF UP: ')
ttl.set_position([.5, 1.05])
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

```python
#for data of districts of UP having highest no. of cases in a month
df2=pd.read_csv('TOP 10 DIS UP.csv')
```

```python
df2.head()
```

```python
#barh graph demonstrating districts having highest no. of confirmed cases

fig = plt.figure()
b = fig.add_axes([0,0,1,1])
dt = [2, 3, 5, 6, 8, 0, 7, 4,1,9]
ttl=plt.title('DISTRICTS HAVING HIGHEST NUMBER OF CONFIRMED CASES ')
ttl.set_position([.5, 1.05])
my_cmap = cm.get_cmap('jet')
my_norm = Normalize(vmin=0, vmax=9)
plt.xlabel('CONFIRMED CASES')
plt.ylabel('DISTRICTS')
b.barh('District','Confirmed',data=df2,color=my_cmap(my_norm(dt)))

plt.show() 
```

```python
#bar graph for comparing the confirmed and recovered cases in district of UP with the district of MP nearer to it

labels = ['Etawah(UP):recovery rate-62.17%', 'Tikamgarh(MP):recovery rate-41.70%']
recv = [309,93]
conf = [497,223]

x = np.arange(len(labels)) # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots(figsize=(7,5))
rects1 = ax.bar(x - width/2,conf, width, label='CONFIRMED')
rects2 = ax.bar(x + width/2, recv, width, label='RECOVERED')

ax.set_ylabel('NO. OF CASES')
ax.set_title('Comparing the confirmed and recovered cases of the district of UP with nearer district of MP')
ttl=ax.title
ttl.set_position([.5, 1.05])
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim([0,800])
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
plt.margins()

plt.show()


```
